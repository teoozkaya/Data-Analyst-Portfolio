## Which categories of bicycles are sold most?

```python
query = '''
    SELECT cat.category_name, 
        SUM(oi.quantity) AS total_sold
    FROM categories as cat
    JOIN products as pt ON pt.category_id = cat.category_id
    JOIN order_items as oi ON oi.product_id = pt.product_id
    GROUP BY cat.category_name
    ORDER BY total_sold DESC;
'''

df = pd.read_sql_query(query, conn)
df
```

## Which store had the most amount of customers and sales?

```python
query = '''
SELECT s.store_name, COUNT(DISTINCT o.customer_id) AS total_customers, COUNT(DISTINCT o.order_id) AS total_orders
FROM stores s
JOIN orders o ON s.store_id = o.store_id
GROUP BY s.store_id
ORDER BY total_customers DESC;
'''

df = pd.read_sql_query(query, conn)
df
```

### Which store has the highest revenue and what are their percentage?

```python
query = ''' 
WITH total_revenue AS (
    SELECT s.store_name, ((oi.list_price * oi.quantity) - (1 - oi.discount)) AS total_sale_product
    FROM  order_items AS oi
    LEFT JOIN orders as o ON o.order_id = oi.order_id
    LEFT JOIN stores as s ON s.store_id = o.store_id)

SELECT store_name, 
    SUM(total_sale_product) AS revenue, 
    ROUND((SUM(total_sale_product) / ((SELECT SUM(total_sale_product) FROM total_revenue)) * 100),2) as percentage
FROM total_revenue
GROUP BY store_name
ORDER BY revenue DESC;

'''
df = pd.read_sql_query(query, conn)
df
```

## Who is the top customer?

```python
query = '''
    SELECT c.customer_id, 
        c.first_name, 
        c.last_name, 
        ROUND(SUM((oi.quantity * oi.list_price) * (1-oi.discount)),2) AS total_spent
    FROM customers AS c
    LEFT JOIN orders AS o ON o.customer_id = c.customer_id
    LEFT JOIN order_items AS oi ON o.order_id = oi.order_id
    GROUP BY c.customer_id, c.first_name, c.last_name
    ORDER BY total_spent DESC
    LIMIT 10;
'''

df = pd.read_sql_query(query, conn)
df
```

## Classifying Customers

```python
query = '''
    WITH customer_stats AS (
        SELECT customer_id,
            SUM((quantity * list_price) * (1-discount)) AS total_spent,
            COUNT(DISTINCT o.order_id) AS total_orders
        FROM orders o
        JOIN order_items AS oi
            ON o.order_id = oi.order_id
        GROUP BY customer_id

    )

    SELECT customer_id,
        CASE WHEN total_orders > 1 THEN 'No'
            ELSE 'Yes'
            END AS one_time_buyer,
        CASE WHEN total_spent / (SELECT MAX(total_spent) FROM customer_stats) > 0.7 THEN 'top spender'
            WHEN total_spent / (SELECT MAX(total_spent) FROM customer_stats) > 0.35 THEN 'average spender'
            ELSE 'low spender'
            END AS spender_position
    FROM customer_stats;
'''
df = pd.read_sql_query(query, conn)
df
```

## Finding the average brand sale for each month

```python
query = '''
    WITH sale_dates AS (
        SELECT strftime('%Y', order_date) AS year,
               strftime('%m', order_date) AS month,
                oi.product_id,
                SUM(quantity) as total_sales_that_month
        FROM orders o
        INNER JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY 1,2,3
    ),
    product_brands AS (
        SELECT p.product_id,
            brand_name
        FROM products p
        INNER JOIN brands b ON b.brand_id = p.brand_id
    )
    SELECT month,
        brand_name,
        AVG(total_sales_that_month) AS average_sold_products
    FROM sale_dates
    INNER JOIN product_brands ON sale_dates.product_id = product_brands.product_id
    GROUP BY 1, 2;
'''
df = pd.read_sql_query(query, conn)
df
```

## Transition matrix of multiple time buyers based on category

```python
query = '''
    WITH transition_matrix AS (
        SELECT
            p1.category_id AS from_category,
            p2.category_id AS to_category,
            COUNT(o1.order_id) AS total_transitions
        FROM
            orders o1
        INNER JOIN orders o2 ON o1.customer_id = o2.customer_id
        INNER JOIN order_items oi1 ON o1.order_id = oi1.order_id
        INNER JOIN order_items oi2 ON o2.order_id = oi2.order_id
        INNER JOIN products p1 ON p1.product_id = oi1.product_id
        INNER JOIN products p2 ON p2.product_id = oi2.product_id
        WHERE o1.order_date < o2.order_date
        GROUP BY 1,2
    ),

    transition_probabilities AS (
        SELECT 
            from_category,
            to_category,
            total_transitions,
            total_transitions * 1.0 / SUM(total_transitions) OVER (PARTITION BY from_category) AS transition_probability
        FROM 
            transition_matrix
    )

    SELECT 
        c1.category_name AS from_category,
        c2.category_name AS to_category,
        total_transitions,
        transition_probability
    FROM 
        transition_probabilities
    INNER JOIN categories c1 ON c1.category_id = transition_probabilities.from_category
    INNER JOIN categories c2 ON c2.category_id = transition_probabilities.to_category
    ORDER BY 1,2
'''

df = pd.read_sql_query(query, conn)
df
```
